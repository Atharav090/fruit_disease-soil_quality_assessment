```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. LOAD THE STRICTLY EXTRACTED DATA
papers_data = [
    {"Model": "[3] J48 and ANN", "Accuracy": 95.0},
    {"Model": "[6] Random Forest", "Accuracy": 98.3},
    {"Model": "[7] Classification and Regression Trees (CART)", "Accuracy": 93.3},
    {"Model": "[10] Back Propagation Network (BPN)", "Accuracy": 94.2},
    {"Model": "[11] Extreme Learning Machine (ELM)", "Accuracy": 83.3},  
    {"Model": "[13] Support Vector Machine (SVM)", "Accuracy": 92},  
    {"Model": "[15] Artificial Neural Networks (ANN)", "Accuracy": 93},  
    {"Model": "[18] Random Forest", "Accuracy": 96.0},
    {"Model": "SoilVisioNet (Ours)", "Accuracy": 98.12}
]

df = pd.DataFrame(papers_data)
df.set_index("Model", inplace=True)

# 2. CONFIGURE PLOT AESTHETICS (Times New Roman, Clear Linings, High Resolution)
plt.rcParams.update({
    'font.family': 'Times New Roman',
    'font.size': 12,
    'axes.edgecolor': 'black',
    'axes.linewidth': 1.2
})

metrics = ["Accuracy"]
colors = ['#4A90E2']

fig, ax = plt.subplots(figsize=(12, 7), dpi=1000)

x = np.arange(len(df.index))
width = 0.5  # Width of each bar

# 3. GENERATE CENTERED BARS WITH BOX LININGS
for i, metric in enumerate(metrics):
    # Set offset to 0 since there is only one metric being plotted
    offset = 0 
    bars = ax.bar(
        x + offset, 
        df[metric], 
        width, 
        label=metric, 
        color=colors[i], 
        edgecolor='black', # Proper linings/boxes
        linewidth=1.2,
        zorder=3
    )
    
    # Tidily write text annotations clearly above each bar
    for bar in bars:
        height = bar.get_height()
        if not np.isnan(height) and height > 0:
            ax.annotate(
                f'{height:.2f}%',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 6),  # Offset slightly higher for neatness
                textcoords="offset points",
                ha='center', va='bottom', 
                fontsize=10, 
                fontweight='bold', 
                rotation=0,
                color='black'
            )

# 4. FORMATTING FOR CLARITY
ax.set_ylabel('Performance Metric (%)', fontweight='bold', fontsize=14)
ax.set_title('Comparative Performance Analysis: Literature vs. SoilVisioNet', fontsize=16, fontweight='bold', pad=20)

# Apply rotation and alignment to prevent overlapping
ax.set_xticks(x)
ax.set_xticklabels(df.index, fontweight='bold', fontsize=12, rotation=45, ha='right')

# Adjust Y-axis limit to give room for tidily written text
min_val = df.min().min()
ax.set_ylim(max(0, min_val - 5), 105) 

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.25), ncol=4, fontsize=12, frameon=True, edgecolor='black')
ax.grid(axis='y', linestyle='--', alpha=0.7, zorder=0)

plt.tight_layout()

# Save at exactly 1000 DPI
plt.savefig('strict_comparative_analysis_1000dpi.png', dpi=1000, bbox_inches='tight')
plt.show()
```
